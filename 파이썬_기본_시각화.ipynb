```
import numpy as np
```

```
t = np.arange(0,12,0.01)
```

```
y = np.sin(t)
z = np.cos(t)
```

```
plt.figure(figsize = (10,6))
plt.plot(t, y, label = 'sin', lw = 5)
plt.plot(t, z, 'r--', label = 'cos', color = 'red', linestyle = 'dashed', marker = 'h')

plt.grid() # 격자선이 사라짐
plt.legend() # 우측 상단 범례
plt.xlabel('time')
plt.ylabel('value')
plt.title('my graph!!')


plt.show()
```

```
t2 = np.arange(0, 5, 0.5)
```

```
plt.figure(figsize = (10,6))
```

```
plt.plot(t2,t2,'r--')
plt.plot(t2,t2*2, 'bs')
plt.plot(t2,t2*3, 'g^')
plt.plot(t2,t2*4, color = 'pink',linestyle='dashed',marker = 'o')


plt.grid() # 격자선이 사라짐
plt.legend() # 우측 상단 범례
plt.xlabel('time')
plt.ylabel('distance')
plt.title('my graph!!')


plt.show()
```

```
s1 = np.random.normal(loc = 0, scale = 1, size = 1000)
s2 = np.random.normal(loc = 5, scale = 0.5, size = 1000) 
s3 = np.random.normal(loc = 10, scale = 2, size = 1000) # 총 3000개의 데이터가 랜덤으로 생성
```

```
plt.figure(figsize = (10,6))
plt.plot(s1, label = 's1')
plt.plot(s2, label = 's2')
plt.plot(s3, label = 's3')
plt.legend()
plt.show()
```

```
plt.figure(figsize = (10,6))
plt.boxplot((s1,s2,s3))


plt.show()
```
